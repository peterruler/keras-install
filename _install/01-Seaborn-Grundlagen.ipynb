## Die Daten

Kontext<br>

Diese Datenbank enthält 76 Attribute, jedoch verwenden alle dazu publizierten Experimente lediglich eine Teilmenge von 14 von diesen. Ausschliesslich die Cleveland-Datenbank wurde bisher von ML-Forschern verwendet. Das Feld "goal" bezeichnet zu das Vorhandensein einer Herzkrankheit bei dem Patienten. Es ist ein Integerwert von 0 (nicht vorhanden) bis 4. 

Columns
* age - Alter in Jahren
* sex - (1 = männlich; 0 = weiblich)
* cp - Typ des Brustschmerzes
* trestbps -  Blutdruck ruhend (in mm Hg bei der Aufnahme ins Krankenhaus)
* cholserum -  cholestoral in mg/dl
* fbs(Nüchternglukose > 120 mg/dl) - (1 = wahr; 0 = falsch)
* restecg -  Ergebnis Elektrokardiogramm ruhend
* thalach -  Maximal erreichte Herzrate
* exang  -  Übungsinduzierte Angina (1 = ja; 0 = nein)
* oldpeak -  ST-Depression durch körperliche Betätigung im Verhältnis zur Ruhe
* slope  -  die Steigung des ST-Segments bei Maximalleistung
* ca  -  Anzahl der durch Flouroskopie gefärbten Hauptgefäße (0-3)
* thal3 = normal; 6 = behobener Defekt; 7 = reversibler Defekt
* target   1 oder 0

## Pairplots

Pairplots beinhalten Scatterplots und Histogramme für jede einzelne Spalte im Datensatz. Bei großen Datensätzen kann das ein riesiger Plot sein! Verwende Pairplots daher mit vorsicht, da es eine lange Zeit dauern kann für große Datensätze und die resultierenden Plots gegebenenfalls zu klein sind!

INFO: https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
iris = pd.read_csv('../DATA/iris.csv')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
# Shows KDEs instead of histograms along the diagonal
sns.pairplot(iris, hue="species")
```
